```python
from matplotlib import pyplot as plt
import numpy as np
from math import log2, sqrt, sin
```

```python
thickness = 2
n = 10**100
small_n = 10**3
```

# O notation

Visualizing asymptotic growth for two functions `f` and `g` as x is growing. We are unable to see what happens after $x = 10^{100}$ because of limitations in Python but for some functions visualizing the growth relationship can help a lot.

### Oppgave 1


```python
def f(x):
    return x**3

def g(x):
    return x**(log2(3))
```

```python
x = np.linspace(1,n)

f2 = np.vectorize(f)
g2 = np.vectorize(g)

plt.plot(x,f2(x), label="f", linewidth=thickness)
plt.plot(x,g2(x), label="g", linewidth=thickness)

plt.legend()

plt.show()
```

### Oppgave 3


$f=\frac{(n^3 + 20)}{n}$ = $n^2 \frac{20}{n}$  
  
$g=\frac{n^3}{(n+20)}$ = $n^2 - 20n \frac{8000}{(20 + n)} + 400$

```python
def f(x):
    return (x**3 + 20) / x

def g(x):
    return x**3 / (x+20)
```

```python
x = np.linspace(1,n)

f2 = np.vectorize(f)
g2 = np.vectorize(g)

plt.plot(x,f2(x), label="f", linewidth=thickness+2)
plt.plot(x,g2(x), label="g", linewidth=2)

plt.legend()

plt.show()
```

### Oppgave 8


```python
def f(x):
    return x + log2(x)

def g(x):
    return 3*x

def l(x):
    return log2(x)
```

```python
x = np.linspace(1,n)

f2 = np.vectorize(f)
g2 = np.vectorize(g)
l2 = np.vectorize(l)

plt.plot(x,f2(x), label="f", linewidth=thickness)
plt.plot(x,g2(x), label="g", linewidth=thickness)
plt.plot(x,l2(x), label="log(n)", linewidth=thickness)


plt.legend()

plt.show()
```

How $log2(n)$ grows

```python
l2 = np.vectorize(l)
plt.plot(x,l2(x), label="log(n)", linewidth=thickness)
plt.legend()
plt.show()
```
